```python
# First, install the required packages
!pip install seaborn pandas matplotlib

# Task 1: Exploring and Visualizing a Simple Dataset (Iris)

# 1. Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Load the dataset (Seaborn has a built-in Iris dataset)
iris = sns.load_dataset("iris")

# 3. Inspect the dataset
print("Shape of dataset:", iris.shape)   # rows x columns
print("\nColumn names:", iris.columns)  # feature names
print("\nFirst 5 rows:\n", iris.head()) # preview data

# 4. Summary information
print("\nDataset Info:")
print(iris.info())

print("\nStatistical Summary:")
print(iris.describe())

# 5. Data Visualization

# Scatter Plot: Relationships between features
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", style="species")
plt.title("Scatter Plot of Sepal Length vs Width")
plt.show()

# Pairplot: Scatter plots between all numerical features
sns.pairplot(iris, hue="species", diag_kind="hist")
plt.suptitle("Pairwise Scatter Plots", y=1.02)
plt.show()

# Histogram: Distribution of features
iris.hist(bins=20, figsize=(10,8), color="skyblue", edgecolor="black")
plt.suptitle("Feature Distributions")
plt.show()

# Boxplot: Identifying outliers
plt.figure(figsize=(8,6))
sns.boxplot(data=iris)
plt.title("Boxplots of Iris Features")
plt.show()
```
